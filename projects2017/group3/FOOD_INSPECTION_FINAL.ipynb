```python
# Get the violation numbers from Violation column
a = insp_data['Violations'].str.extractall('(\d+\.)')
a = a[~a[0].astype(str).str.startswith('0')]
a = a.reset_index(level=1, drop=True)

# replace .
a = a[0]
a = a.str.replace(r'\.','')
```

```python
def remove_duplicates(values):
    '''
    Function to remove duplicates within the list of violations
    '''
    output = []
    seen = set()
    for i in values:
        if i not in seen:
            output.append(i)
            seen.add(i)
    return output
```

```python
# Get the violation list, remove duplicates and put them into an array 
v = []
ind = []
count = []
index = insp_data.index
for i in range(len(index)):
    try:
        b = np.array(a[index[i]])
        if len(b) > 1:
            violations = remove_duplicates(b)   
            v.append(violations)
        ind.append(index[i])
        count.append(len(violations))
    except TypeError:
        v.append(a[index[i]])
        ind.append(index[i])
        count.append(1)
        continue
    except:
        v.append(0)
        ind.append(index[i])
        count.append(0)
        continue
```

```python
vio = pd.DataFrame({'Violations':v})
vio = np.asarray(vio['Violations'])
vio = pd.DataFrame({'Violations':vio})
vio['Index'] = ind
vio['No. of Violations'] = count
```
