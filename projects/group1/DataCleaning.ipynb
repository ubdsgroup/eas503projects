```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time

data = pd.read_csv('Energy_Usage_2010.csv')
df = pd.DataFrame(data)
df=data.sort_values(['COMMUNITY AREA NAME'])

# Check Shape
print(df.shape)

# Check total NaN count
print("Total NaN's: ",df.isnull().sum().sum())

# Drop unnecessary columns
df2=df.drop(df.columns.to_series()["KWH MEAN 2010":"THERMS SQFT MAXIMUM 2010"],axis=1)
df2=df2.drop("OCCUPIED HOUSING UNITS",axis=1)

# Create Community Area DataFrame to store aggregate Community Area Data, then remove data missing from this.
CMA = df2.loc[df['CENSUS BLOCK'].isnull()]
CMA = CMA.drop(["CENSUS BLOCK","BUILDING TYPE","BUILDING_SUBTYPE"],axis=1)

# Remove 77 rows from main dataframe that were moved to the above CBDF and rows missing KWH data and Total Therm data
df3 = df2.dropna(subset=["CENSUS BLOCK","KWH JANUARY 2010","TOTAL THERMS"], thresh=3)

# Check shape after cleaning
print(df3.shape)

# Check total NaN count after cleaning
print("Total NaN's: ",df3.isnull().sum().sum())

# See what's still missing, we elected to keep the columns missing some Therm data for now
print(df3.isnull().sum())

# Define each community area in each of the 9 districts
FarNorth = ["FarNorth","Rogers Park","West Ridge","Uptown","Lincoln Square","Edison Park","Norwood Park","Jefferson Park","Forest Glen","North Park","Albany Park","O'Hare","Edgewater"]
North = ["North","North Center","Lakeview","Lincoln Park","Avondale","Logan Square"]
Northwest = ["Northwest","Portage Park","Irving Park","Dunning","Montclare","Belmont Cragin","Hermosa"]
Central =["Central","Near North Side","Loop","Near South Side"]
West =["West","Humboldt Park","West Town","Austin","West Garfield Park","East Garfield Park","Near West Side","North Lawndale","South Lawndale","Lower West Side"]
Southwest =["Southwest","Garfield Ridge","Archer Heights","Brighton Park","McKinley Park","New City","West Elsdon","Gage Park","Clearing","West Lawn","Chicago Lawn","West Englewood","Englewood"]
South =["South","Armour Square","Douglas","Oakland","Fuller Park","Grand Boulevard","Kenwood","Washington Park","Hyde Park","Woodlawn","South Shore","Bridgeport","Greater Grand Crossing"]
FarSouthwest=["FarSouthwest","Ashburn","Auburn Gresham","Beverly","Washington Heights","Mount Greenwood","Morgan Park"]
FarSouth =["FarSouth","Chatham","Avalon Park","South Chicago","Burnside","Calumet Heights","Roseland","Pullman","South Deering","East Side","West Pullman","Riverdale","Hegewisch"]

# Concatenate all district lists into single list, initialize empty "dist" data frame 
districts = [FarNorth,North,Northwest,Central,West,Southwest,South,FarSouthwest,FarSouth]
dist=pd.DataFrame(columns=CMA.columns.values)

# Iterate through the districts list and filter the cleaned "df3" data frame to select only sums of requested community areas
for district in districts:
    part = CMA[CMA['COMMUNITY AREA NAME'].isin(district)]
    distsum=part.sum(axis=0)
    dist=dist.append(distsum,ignore_index=True)

# Rename "community area names" to "district"
dist=dist.rename(columns = {'COMMUNITY AREA NAME':'DISTRICT'})

# Apply district names, normalize columns that were averages
i=0
for district in districts:
    dist.iloc[i,0]=districts[i][0]
    dist.iloc[i,[34,35,36,38,40]]=dist.iloc[i,[34,35,36,38,40]]/len(district)
    i+=1

#Export cleaned Census Block, Community Area and District df's back to .csv's
df3.to_csv("Census_Blocks.csv")
CMA.to_csv("Community_Areas.csv")
dist.to_csv("Districts.csv")
```
